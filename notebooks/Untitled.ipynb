```python
la = pd.read_hdf('/data/jls/GaiaDR2/spectro/LAMOST_cannon_sample.hdf5')
```

```python
a = pd.read_hdf('/data/jls/GaiaDR2/spectro/APOGEE_input.hdf5')
```

```python
lamost = pd.read_hdf('/data/jls/GaiaDR2/spectro/LAMOST_input.hdf5')
```

```python
la = pd.merge(la,lamost,on='obsid')
```

```python
la = pd.merge(la,a,on='APOGEE_ID')
```

```python
plt.hist((la.TEFF-la.teff_x)/np.sqrt(la.e_teff_x**2+la.e_teff_y**2),
         range=[-3.,3.],bins=50,normed=True,histtype='step');
xx = np.linspace(-3.,3.)
plt.plot(xx,np.exp(-xx**2/2.)/np.sqrt(2.*np.pi))
```

```python
plt.hist((la.LOGG-la.logg_x+0.15)/np.sqrt(la.e_logg_x**2+la.e_logg_y**2),
         range=[-3.,3.],bins=50,normed=True,histtype='step');
xx = np.linspace(-3.,3.)
plt.plot(xx,np.exp(-xx**2/2.)/np.sqrt(2.*np.pi))
```
